```python
##############################################################################
# Imports
##############################################################################
from utils.graphrag_utils import DependencyAnalyzer
from dotenv import load_dotenv
import logging
import os
from IPython.display import display, Markdown
logging.basicConfig(level=logging.INFO)
load_dotenv()
import asyncio
import nest_asyncio
nest_asyncio.apply()
```

### GraphRAG Queries

```python
##############################################################################
# Instance Variables
##############################################################################
_GRAPHRAG_SOURCE_PATH = f"{os.path.abspath('graph_rag_app')}/source"
analyzer = DependencyAnalyzer(_GRAPHRAG_SOURCE_PATH)
```

#### Adhoc Queries (Perform ad-hoc queries against the GraphRAG index)

```python
##############################################################################
# Sample Questions:
# 1. Which modules would be riskiest to refactor first? Include the fully qualified names.
# 2. What migration order would be recommended when refactoring to reduce breaking changes? Include the fully qualified names.
# 3. Which modules would be the least risky to migrate first? Include the fully qualified names.
# 4. Are there any vulnerable dependencies or libraries? Include the fully qualified names.
##############################################################################
question = """
Which modules would be riskiest to refactor first?
Include the fully qualified names.
"""
response = await analyzer.query_with_llm(question)
display(Markdown(response))
```

#### Generate a migration report

```python
migration_report = await analyzer.generate_migration_report()
display(Markdown(migration_report))
```
